```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Lenovo/Documents/Especialización Finanzas/Semestre 2/Machine Learning/realestate/RDC_Inventory_Core_Metrics_Zip_History.csv')
print('Dataset has ',df.shape[0],' records and ',df.shape[1], ' columns' )
print(' ')
df
```

```python
## Limpieza de la Data
```

```python
df.drop(df.tail(1).index,inplace=True) 
```

```python
df.info()
```

```python
## arreglamos las variables de tipo objeto
```

```python
df = df.drop(['flag'], axis=1)
df.rename(columns={'postal_code': 'ZIP'}, inplace=True)
df['ZIP'] = df['ZIP'].astype('str') 
df['ZIP'] = df['ZIP'].apply(lambda x: x.zfill(5))
```

```python
## corregir el formato fecha
df['month_date_yyyymm'] = pd.to_datetime(df['month_date_yyyymm'],format = '%Y%m')
df.rename(columns={'month_date_yyyymm':'Date'}, inplace=True)
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Date'] = df['Date'].dt.date
df.head()
```

```python
## crear variables de Ciudad y Estado
df[['city','STATE_ABBR']] = df.zip_name.str.split(',',expand=True)
df['STATE_ABBR'] = df['STATE_ABBR'].str.lstrip().str.upper()
df['city'] = df['city'].str.title()

df = df.sort_values(["STATE_ABBR", "Date"], ascending = (True, True)).reset_index(drop = True) # Default is ascending
df = df[ ['STATE_ABBR'] + [ col for col in df.columns if col != 'STATE_ABBR' ] ]
df
```

```python
df['price_increased_count']
```

```python
df.isnull().sum()
```

```python
## eliminar los cambios anuales
df = df.loc[:,~df.columns.str.contains('yy$')]

## eliminar price_increased_count_mm y price_reduced_count_mm por NAs
df = df.loc[:,~df.columns.str.contains('increased_count_mm|median_square_feet_mm|median_listing_price_per_square_foot_mm|pending_listing_count_mm|reduced_count_mm|new_listing_count_mm' )]

## eliminar NAs de mi Y
df.dropna(subset=['median_listing_price_mm','active_listing_count_mm','median_days_on_market_mm'],inplace=True)
```
